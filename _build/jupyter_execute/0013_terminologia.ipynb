# <font color='navy'>  Terminológia  </font> 

Terminológia a použité symboly

| Symbol     | Význam 
|:---        | :--    
| #          | komentár
| ;          | oddelovač 
| 123.45     | hodnota
| x          | premenná
| (x,y)      | súradnica
| {...}      | vetva
| [...]      | blok
| "txt"      | textový reťazec
| `txt'      | textový reťazec v makroprcesore m4
| \$ txt \$   | LaTeX matematický výraz v texte
| meno ...   | inštrukcia
| A:         | referencia na objekt
| *.a*       | atribút objektu
| **P**      | parameter objektu
| meno()     | makro
| $1         | argument makra
